__BA1K__ <br>
__Generate the Frequency Array of a String__

```python
from itertools import product
from Bio.Seq import Seq
def solve(main,t):
  val="ACGT"
  mylist=[''.join(p) for p in product(val,repeat=t)]
  ans=[]
  sequence=Seq(main)

  for i in mylist:
    ans.append(sequence.count_overlap(i))
  return ans
solve("ACGCGGCTCTGAAA",2)
```

__BA1L__ <br>
__Implement PatternToNumber__

```python
def solve(main):
    result=0
    dict={'A':0,'C':1,'G':2,'T':3}
    for i in main:
        result*=4
        result+=dict[i]
    return result
solve("AGT")
```

__BA1M__ <br>
__Implement NumberToPattern__

```python
def solve(val,n):
    result=""
    dict={0:'A',1:'C',2:'G',3:'T'}
    for i in range(n):
        result+=dict[val%4]
        val=val//4
    return result[::-1]
solve(11,3)
```

```python
def hamming_distance(first, second):
    return sum(c1 != c2 for c1, c2 in zip(first, second))

def all_neighbour(k):
    base=['A','C','G','T']
    array=base
    for n in range(k-1):
         array=[i+j for i in array for j in base]

    return array
def solve(main,d):
    base=['A','C','G','T']
    array=[''.join(p) for p in product(base,repeat=len(main))]
    mylist=[]
    for i in array:
        if hamming_distance(i,main)<=d:
            mylist.append(i)
    return mylist
solve("ACG",1)
```
